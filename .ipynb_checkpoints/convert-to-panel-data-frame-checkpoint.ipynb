```python
import xlrd
import xlsxwriter
```

```python
#open excel file
fund_flow = xlrd.open_workbook('fund_flow_lux.xlsx')
fund_return = xlrd.open_workbook('fund_return_lux.xlsx')
fund_size = xlrd.open_workbook('fund_size_luxembourg20200108.xlsx')
fund_snapshot = xlrd.open_workbook('fund_snapshot_lux.xlsx')
fund_sustainability = xlrd.open_workbook('fund_sustainability_lux.xlsx')

table_flow = fund_flow.sheets()[0]
table_return = fund_return.sheets()[0]
table_size = fund_size.sheets()[0]
table_snapshot = fund_snapshot.sheets()[0]
table_sustainability = fund_sustainability.sheets()[0]
```

```python
for i in range(0,table_flow.nrows):
    locals()['flow_'+str(i)] = table_flow.row_values(i)
    locals()['flow_month_'+str(i)] = locals()['flow_'+str(i)][7:]
    #共3703个flow_i,flow_i表示该行信息
    
    
for i in range(0,table_return.nrows):
    locals()['return_'+str(i)] = table_return.row_values(i)
    locals()['return_month_'+str(i)] = locals()['return_'+str(i)][7:]
    #共3707个return_i,return_i表示该行信息
    
    
for i in range(0,table_size.nrows):
    locals()['size_'+str(i)] = table_size.row_values(i)
    locals()['size_month_'+str(i)] = locals()['size_'+str(i)][7:]
    #共8001个size_i,size_i表示该行信息
    
for i in range(0,table_sustainability.nrows):
    locals()['sustainability_'+str(i)] = table_sustainability.row_values(i)
    locals()['sustainability_month_'+str(i)] = locals()['sustainability_'+str(i)][5:]
    #共3702个sustainability_i,sustainability_i表示该行信息
```

```python
#flow
flow_FundID = []
flow_FundCode = []
flow_FundLegalName = []
flow_Name = []
flow_BaseCurrency = []
flow_ISIN = []
flow_Domicile = []
flow_monthValue = []

for i in range(0,table_flow.nrows):
    flow_FundID.append(locals()['flow_'+str(i)][0])
    flow_FundCode.append(locals()['flow_'+str(i)][1])
    flow_FundLegalName.append(locals()['flow_'+str(i)][2])
    flow_Name.append(locals()['flow_'+str(i)][3])
    flow_BaseCurrency.append(locals()['flow_'+str(i)][4])
    flow_ISIN.append(locals()['flow_'+str(i)][5])
    flow_Domicile.append(locals()['flow_'+str(i)][6])
    flow_monthValue.append(locals()['flow_'+str(i)][7:101])
    
    
#return
return_FundID = []
return_FundCode = []
return_FundLegalName = []
return_Name = []
return_BaseCurrency = []
return_ISIN = []
return_Domicile = []
return_monthValue = []

for i in range(0,table_return.nrows):
    return_FundID.append(locals()['return_'+str(i)][0])
    return_FundCode.append(locals()['return_'+str(i)][1])
    return_FundLegalName.append(locals()['return_'+str(i)][2])
    return_Name.append(locals()['return_'+str(i)][3])
    return_BaseCurrency.append(locals()['return_'+str(i)][4])
    return_ISIN.append(locals()['return_'+str(i)][5])
    return_Domicile.append(locals()['return_'+str(i)][6])
    return_monthValue.append(locals()['return_'+str(i)][7:101])
    
    
#size
size_FundID = []
size_FundCode = []
size_FundLegalName = []
size_Name = []
size_BaseCurrency = []
size_ISIN = []
size_Domicile = []
size_monthValue = []

for i in range(0,table_size.nrows):
    size_FundID.append(locals()['size_'+str(i)][0])
    size_FundCode.append(locals()['size_'+str(i)][1])
    size_FundLegalName.append(locals()['size_'+str(i)][2])
    size_Name.append(locals()['size_'+str(i)][3])
    size_BaseCurrency.append(locals()['size_'+str(i)][4])
    size_ISIN.append(locals()['size_'+str(i)][5])
    size_Domicile.append(locals()['size_'+str(i)][6])
    size_monthValue.append(locals()['size_'+str(i)][7:101])
    
'''
#sustainability
sustainability_FundID = []
#sustainability_FundCode = []
#sustainability_FundLegalName = []
sustainability_Name = []
sustainability_BaseCurrency = []
sustainability_ISIN = []
sustainability_Domicile = []
sustainability_monthValue = []

for i in range(0,table_sustainability.nrows):
    sustainability_FundID.append(locals()['sustainability_'+str(i)][0])
    sustainability_FundCode.append(locals()['sustainability_'+str(i)][1])
    sustainability_FundLegalName.append(locals()['sustainability_'+str(i)][2])
    sustainability_Name.append(locals()['sustainability_'+str(i)][3])
    sustainability_BaseCurrency.append(locals()['sustainability_'+str(i)][4])
    sustainability_ISIN.append(locals()['sustainability_'+str(i)][5])
    sustainability_Domicile.append(locals()['sustainability_'+str(i)][6])
    sustainability_monthValue.append(locals()['sustainability_'+str(i)][7:101])
''' 

#build a list of month to write in the excel file   
month_list = []
for i in range(1,95):
    if i <= 12:
        month = "2012"+"-"+str(i)
        month_list.append(month)
    elif i > 12:
        if i%12 == 0:
            month = str(2012+(i//12)-1)+"-"+str(12)
            month_list.append(month)
        else:
            month = str(2012+(i//12))+"-"+str(i%12)
            month_list.append(month)
```

```python
#Fund ID, which are in the Fund flow but not in the Fund size
common_size_flow = []

for i in range(1,len(flow_FundID)):
    for j in range(1,len(size_FundID)):
        if flow_FundID[i]==size_FundID[j]:
            common_size_flow.append(flow_FundID[i])
                
in_flow_notIn_size = [x for x in flow_FundID if x not in common_size_flow]
print("Fund ID which are only in the flow but not in the size")
print(in_flow_notIn_size)

#Fund ID, which are in the Fund return but not in the Fund size
common_size_return = []

for i in range(1,len(return_FundID)):
    for j in range(1,len(size_FundID)):
        if return_FundID[i]==size_FundID[j]:
            common_size_return.append(return_FundID[i])
                
in_return_notIn_size = [x for x in return_FundID if x not in common_size_return]
print("Fund ID which are only in the return but not in the size")
print(in_return_notIn_size)
```

```python
workbook = xlsxwriter.Workbook('Summary_export.xlsx')
worksheet = workbook.add_worksheet()

#write basic data of flow into the file
for i in range(1,table_size.nrows):
    #Fund ID in the 0 col
    worksheet.write(0, 0, locals()['size_'+str(0)][0])
    for j in range(0,94):
        worksheet.write(j+94*(i-1)+1, 0, locals()['size_'+str(i)][0])
    
    #Fund Code in the 1 col
    worksheet.write(0, 1, locals()['size_'+str(0)][1])
    for j in range(0,94):
        worksheet.write(j+94*(i-1)+1, 1, locals()['size_'+str(i)][1])

    #Fund Legal Name in the 1 col
    worksheet.write(0, 2, locals()['size_'+str(0)][2])
    for j in range(0,94):
        worksheet.write(j+94*(i-1)+1, 2, locals()['size_'+str(i)][2])
    
    #Name in the 1 col
    worksheet.write(0, 3, locals()['size_'+str(0)][3])
    for j in range(0,94):
        worksheet.write(j+94*(i-1)+1, 3, locals()['size_'+str(i)][3])
    
    #BaseCurrency in the 1 col
    worksheet.write(0, 4, locals()['size_'+str(0)][4])
    for j in range(0,94):
        worksheet.write(j+94*(i-1)+1, 4, locals()['size_'+str(i)][4])
    
    #ISIN in the 1 col
    worksheet.write(0, 5, locals()['size_'+str(0)][5])
    for j in range(0,94):
        worksheet.write(j+94*(i-1)+1, 5, locals()['size_'+str(i)][5])
    
    #Domicile in the 1 col
    worksheet.write(0, 6, locals()['size_'+str(0)][6])
    for j in range(0,94):
        worksheet.write(j+94*(i-1)+1, 6, locals()['size_'+str(i)][6])
    
    
#write month into the file
worksheet.write(0, 7, "Month")
for i in range(1,table_size.nrows):
    for j in range(0,94):
        worksheet.write(j+94*(i-1)+1, 7, month_list[j])
        
        
#Value of Fund Size for each fund fill into the file
worksheet.write(0, 8, "Fund Size")
for i in range(1,table_size.nrows):
    for j in range(0,94):
        worksheet.write(j+94*(i-1)+1, 8, size_monthValue[i][j])
        
        
#search for the correspondence Fund ID and write Fund flow in the table
worksheet.write(0, 9, "Fund Flow")
for i in range(1,len(flow_FundID)):
    for j in range(1,len(size_FundID)):
        if flow_FundID[i]==size_FundID[j]:
            #position: col 9, row j+94*(i-1)+1
            for z in range(0,94):
                worksheet.write(z+94*(j-1)+1, 9, flow_monthValue[i][z])
                
                
#search for the correspondence Fund ID and write Fund return in the table
worksheet.write(0, 10, "Fund Return")
for i in range(1,len(return_FundID)):
    for j in range(1,len(size_FundID)):
        if return_FundID[i]==size_FundID[j]:
            #position: col 10, row j+94*(i-1)+1
            for z in range(0,94):
                worksheet.write(z+94*(j-1)+1, 10, return_monthValue[i][z])
                
workbook.close()
```
