```python
import xlrd
import xlsxwriter
```

```python
#open excel file
fund_size = xlrd.open_workbook('fund_size_luxembourg20200108.xlsx')
table_size = fund_size.sheets()[0]
```

```python
for i in range(0,table_size.nrows):
    locals()['size_'+str(i)] = table_size.row_values(i)
    locals()['size_month_'+str(i)] = locals()['size_'+str(i)][7:]
    
    #共8001个size_i,size_i表示该行信息
```

```python
size_not_empty = []

for i in range(0,table_size.nrows):
    for j in range(0,len(locals()['size_month_'+str(i)])):
        if locals()['size_month_'+str(i)][j] != '':
            size_not_empty.append(i)
            break
        else:
            continue
            
#find all empty value (row index) of fund_size, totoally 5754
```

```python
size_FundID = []
size_FundCode = []
size_FundLegalName = []
size_Name = []
size_BaseCurrency = []
size_ISIN = []
size_Domicile = []
size_monthValue = []

for i in size_not_empty:
    size_FundID.append(locals()['size_'+str(i)][0])
    size_FundCode.append(locals()['size_'+str(i)][1])
    size_FundLegalName.append(locals()['size_'+str(i)][2])
    size_Name.append(locals()['size_'+str(i)][3])
    size_BaseCurrency.append(locals()['size_'+str(i)][4])
    size_ISIN.append(locals()['size_'+str(i)][5])
    size_Domicile.append(locals()['size_'+str(i)][6])
    size_monthValue.append(locals()['size_'+str(i)][7:101])
    
#all properties are in range of [size_not_empty]
```

```python
workbook = xlsxwriter.Workbook('size_empty_removed.xlsx')
worksheet = workbook.add_worksheet()

worksheet.write_column(0, 0, size_FundID)
worksheet.write_column(0, 1, size_FundCode)
worksheet.write_column(0, 2, size_FundLegalName)
worksheet.write_column(0, 3, size_Name)
worksheet.write_column(0, 4, size_BaseCurrency)
worksheet.write_column(0, 5, size_ISIN)
worksheet.write_column(0, 6, size_Domicile)

for row in range(len(size_monthValue)):
    worksheet.write_row(row, 7, size_monthValue[row])
    
workbook.close()
```
