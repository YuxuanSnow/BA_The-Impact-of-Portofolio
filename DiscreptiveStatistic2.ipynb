```python
import numpy as np
import pandas as pd
import pylab as pl
import matplotlib.pyplot as plt
import scipy
from scipy.stats.mstats import winsorize
from scipy.stats import skew
from scipy.stats import kurtosis
from scipy import stats
```

```python
discreptive_capm = pd.read_excel("AfterRegression.xlsx")
```

```python
#for t test: Performance
siz_dropna_perf = discreptive_capm[['CAPM Performance', 'Performance Three Factor', 'Performance Four Factor','Fund Size']].dropna()
#sus_dropna.dropna(subset=["Fund Sustainability"], inplace=True)

siz_med_perf = siz_dropna_perf["Fund Size"].median()
low_siz_perf = siz_dropna_perf[siz_dropna_perf['Fund Size'] < siz_med_perf]
high_siz_perf = siz_dropna_perf[siz_dropna_perf['Fund Size'] > siz_med_perf]
```

```python
#for t test: Performance
siz_dropna_esg = discreptive_capm[['Fund Sustainability','Fund Size']].dropna()

siz_med_esg = siz_dropna_esg["Fund Size"].median()
low_siz_esg = siz_dropna_esg[siz_dropna_esg['Fund Size'] < siz_med_esg]
high_siz_esg = siz_dropna_esg[siz_dropna_esg['Fund Size'] > siz_med_esg]
```

```python
def t_test(mean_high, std_high, mean_low, std_low, amount_high, amount_low):
    
    rvs_high = stats.norm.rvs(loc=mean_high, scale=std_high, size=amount_high)
    rvs_low = stats.norm.rvs(loc=mean_low, scale=std_low, size=amount_low)
    
    t_test = stats.ttest_ind(rvs_high, rvs_low, equal_var = False)
    
    return t_test
```

```python
t_test_capm = t_test(high_siz_perf['CAPM Performance'].mean(), high_siz_perf['CAPM Performance'].std(), low_siz_perf['CAPM Performance'].mean(), low_siz_perf['CAPM Performance'].std(), len(high_siz_perf), len(low_siz_perf))
t_test_performance_3 = t_test(high_siz_perf['Performance Three Factor'].mean(), high_siz_perf['Performance Three Factor'].std(), low_siz_perf['Performance Three Factor'].mean(), low_siz_perf['Performance Three Factor'].std(), len(high_siz_perf), len(low_siz_perf))
t_test_performance_4 = t_test(high_siz_perf['Performance Four Factor'].mean(), high_siz_perf['Performance Four Factor'].std(), low_siz_perf['Performance Four Factor'].mean(), low_siz_perf['Performance Four Factor'].std(), len(high_siz_perf), len(low_siz_perf))
t_test_esg = t_test(high_siz_esg['Fund Sustainability'].mean(), high_siz_esg['Fund Sustainability'].std(), low_siz_esg['Fund Sustainability'].mean(), low_siz_esg['Fund Sustainability'].std(), len(high_siz_esg), len(low_siz_esg))
```
