```python
import numpy as np
import pandas as pd
```

```python
summary_notclean = pd.read_excel('Summary_export.xlsx')
table_OEF = summary_notclean[summary_notclean['Investment Type'] == 'Open-End Fund']
table_OEF.to_excel("Open_End Fund.xlsx")
#read the summary, firsty clean the Investment Type, only Open-End Fund reamin
```

```python
table_OEF = table_OEF.fillna(0)
#set all NAN to 0, which sppeds up the filtering
```

```python
list_FundSize_Which = []
list_FundSize_AbWann = []
for i in range(3379):
    for j in range(92):
        if abs(table_OEF.iloc[i*92+j]['Fund Size']) > 5000000:
            list_FundSize_Which.append(i)
            list_FundSize_AbWann.append(j)
            break
```

```python
list_FundFlow_Which = []
for i in range(3379):
    for j in range(92):
        if table_OEF.iloc[i*92+j]['Fund Flow'] != 0:
            list_FundFlow_Which.append(i)
            break
```

```python
list_FundReturn_Which = []
for i in range(3379):
    for j in range(92):
        if table_OEF.iloc[i*92+j]['Fund Return'] != 0:
            list_FundReturn_Which.append(i)
            break
```

```python
list_FundSus_Which = []
for i in range(3379):
    for j in range(92):
        if table_OEF.iloc[i*92+j]['Fund Sustainability'] != 0:
            list_FundSus_Which.append(i)
            break
```

```python
#get the index of rows which are not usable
list_Elements = range(3379)

size_NotUsable = [x for x in list_Elements if x not in list_FundSize_Which]
flow_NotUsable = [x for x in list_Elements if x not in list_FundFlow_Which]
return_NotUsable = [x for x in list_Elements if x not in list_FundReturn_Which]
sus_NotUsable = [x for x in list_Elements if x not in list_FundSus_Which]


#get the FundID which are not usable
fundID_size_NotUsable = []
fundID_flow_NotUsable = []
fundID_return_NotUsable = []
fundID_sus_NotUsable = []


#list of index to list of FundID
for i in size_NotUsable:
    fundID_size_NotUsable.append(table_OEF.iloc[i*92]['FundId'])
    
for i in flow_NotUsable:
    fundID_flow_NotUsable.append(table_OEF.iloc[i*92]['FundId'])
    
for i in return_NotUsable:
    fundID_return_NotUsable.append(table_OEF.iloc[i*92]['FundId'])
    
for i in sus_NotUsable:
    fundID_sus_NotUsable.append(table_OEF.iloc[i*92]['FundId'])
    
    
#clean up the FundId rows of the lists
for i in range(len(fundID_size_NotUsable)):
    table_OEF = table_OEF.drop(table_OEF[table_OEF.FundId == fundID_size_NotUsable[i]].index)

for i in range(len(fundID_flow_NotUsable)):
    table_OEF = table_OEF.drop(table_OEF[table_OEF.FundId == fundID_flow_NotUsable[i]].index)
    
for i in range(len(fundID_return_NotUsable)):
    table_OEF = table_OEF.drop(table_OEF[table_OEF.FundId == fundID_return_NotUsable[i]].index)
    
for i in range(len(fundID_sus_NotUsable)):
    table_OEF = table_OEF.drop(table_OEF[table_OEF.FundId == fundID_sus_NotUsable[i]].index)
```

```python
table_OEF.to_excel("Summary Cleaned.xlsx")
```
